```python
import seaborn as sns
import pandas as pd
```

### For CO2

```python
mol ='CO2'
mol_path = f'../{mol}'
df_param_ex = pd.read_csv(f'{mol_path}/exact_param_{mol}.csv')
df_param_ap = pd.read_csv(f'{mol_path}/approx_param_{mol}.csv')
```

```python
heatmap_plot = sns.heatmap(df_param_ap.corr())
fig = heatmap_plot.get_figure()
fig.savefig(f"../images/{mol}_ap.png") 
```

```python
sns.heatmap(df_param_ex.corr())
fig = heatmap_plot.get_figure()
fig.savefig(f"../images/{mol}_ex.png")
```
